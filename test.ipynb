```python
import serial
import time
import matplotlib.pyplot as plt
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_data_from_car():
    ser= serial.Serial("/dev/ttyUSB0", 115200, timeout=1)
    time.sleep(1)
    data = []
    try:
        while True:
            line= ser.readline()
            if line:
                string= line.decode()
                print(string)
                data.append(string)
    except KeyboardInterrupt:
        print("closed")
        ser.close()
    return [list(map(int,re.findall(r'\d+',s))) for s in data]
```

```python
A=get_data_from_car()
```

```python
#LEFT 90 degrees turn
def get_the_outputs_for_plot(data):
    left = [s[0] for s in data]
    right = [s[1] for s in data]
    error = [s[2] for s in data]
    return left,right,error
```

```python
curved_turn = A[482:-541] # handpicked selection of the start and end point of the race
```

```python
def visualize(rightmotor,leftmotor,error):
    fig, (ax1,ax2)= plt.subplots(2)
    timeser=[i*0.005 for i in range(len(leftmotor))] # 0.01 sec delay
    width1 = 12
    height1 = 3
    width_height_1 = (width1, height1)
#     ax1.figure(figsize=width_height_1)
#     ax2.figure(figsize=width_height_1)

    lm_plot= ax1.plot(timeser,leftmotor, label='left motor')
    rm_plot= ax1.plot(timeser,rightmotor, label='right motor')
    ax1.legend(loc='upper center', bbox_to_anchor=(1.15, 1.05),
          ncol=1, fancybox=True, shadow=True)
        
    ax2.plot(timeser,error, label='error')
    ax2.legend(loc='upper center', bbox_to_anchor=(1.1, 1.05),
          ncol=1, fancybox=True, shadow=True)
    
    plt.xlabel("Time (s)")
    plt.setp(ax1, ylabel="Motor's signal (Speed)")
    plt.setp(ax2, ylabel="Error")
    
    
    
```

```python
# Curved turn of the car robot results
# return all the essential data for ploting 
leftmotor, rightmotor, error = get_the_outputs_for_plot(curved_turn)
visualize(leftmotor, rightmotor, error)
```

```python
def clear_data(data):
    processed=[]
    for i in range(1,len(data)):
        if data[i][0]!=0:
            processed.append(data[i-1])
    return processed
processed_straight_line = clear_data(straight_line)
```
